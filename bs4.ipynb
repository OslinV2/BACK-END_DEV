```python
import requests
from bs4 import BeautifulSoup

html_page = 'https://en.wikipedia.org/wiki/Hanumankind'
response = requests.get(html_page, verify=False)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "lxml") # Ou html.parser
    print(soup.prettify())
else:
    print(f"Erro ao acessar a página. Status Code: {response.status_code}")
```

```python
# Algumas formas de navegar pelo site
soup.title
soup.find_all('a')
soup.find_all('b')
soup.find_all_previous('p')
soup.find(id="link3")
soup.title
soup.title.name
soup.title.string
soup.title.parent.name
soup.p
soup.p['class']
soup.a
soup.find_all('a')
soup.find(id="link3")
print(soup.get_text())
```
